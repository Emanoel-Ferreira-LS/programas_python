```python
import matplotlib
```

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples=200, n_features=1,noise=30)
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.show()
```

-importamos a biblioteca matplotlib
-geramos um método de regressão pelo sklearn para gerar as amostras iniciais
-defini o que seria a regressão gerada: 200 exemplos, com caraceristicas envolvendo 1 unidade e uma taxa de erro de 30
-exibi em um grafico
